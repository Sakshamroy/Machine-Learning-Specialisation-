```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# X_trian is input data 
# Y_train is target values 
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m = no of examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0] # len(x_train)
print(f"number of training examples {m}")
```

```python
i = 0# show data values at i = 0 
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
# plotting function using matplotlib 
plt.scatter(x_train, y_train, marker='o',c='b')# scatter use to plot the data points 
plt.title("Housing Price")# give title of the plot
plt.ylabel("price ( in 1000s of dollars)")
plt.xlabel("size ( 1000 sqft)")
plt.show
```

```python
# different values of w and b give different values of st.line 
# fw,b(X(i)) = WX(i) + b
# w = 100, b = 100 
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
# ndarry- show n-dimensional arra of shape(m)
# we need to apply fw,b formula to calucate this we can do individual but doing by for loop will be much better
def compute_model_output(x, w, b):
    """
    we are computing the prediction of a linear model 
    Args:
        x(ndarray(m,)): input data and m examples 
        w,b(scalar): model parameters 
    Returns
        y (ndarray(m,)): target values
    """
    m = x.shape[0]#finding len of examples
    f_wb = np.zeros(m)
    for i in range(m):#creating for loop so that each examples run a time 
        f_wb[i] = w * x[i] + b# formula for calculating 
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
# plot our model prediction 
plt.plot(x_train, tmp_f_wb, c='b', label='Our Prediction')

#  plot the data points
plt.scatter(x_train, y_train, marker='x', c='r', label='Actual Values')

#set the title
plt.title("Housing Prices")

#set y-axis label
plt.ylabel('Price (in 1000s of dollars)')

#set x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()#print discription like our prediction or actual values 
plt.show()
```
